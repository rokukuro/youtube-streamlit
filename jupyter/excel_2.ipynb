```python
export_file_path = 'D:\MyPython\programs\jupyter'
import_folder_path = 'D:\\MyPython\\programs\\jupyter\\rawdate'
path = import_folder_path + '/' + '*.xlsx'
# *.xlsxは全てのxlsxという意味
```

```python
file_path = glob.glob(path)
```

```python
file_path
```

```python
#空のデータフレームを作成
df_concat = pd.DataFrame()

for i in file_path:
  #変数df_read_excel に読み込んだデータを格納
  df_read_excel = pd.read_excel(i)
  #それぞれのデータフレームをくっつける
  df_concat = pd.concat([df_read_excel, df_concat])
```

```python
df_concat
```

```python
#不要なunnamedを削除する
df_drop = df_concat.drop('Unnamed: 0', axis=1)
#列を削除する場合はaxis=1
```

```python
df_drop.head(3)
```

```python
#dfを降順(False)で並べ替える
df_sort = df_drop.sort_values(by='達成率', ascending=False)
```

```python
df_sort
```

```python
#エクスポートする
df_sort.to_excel(export_file_path + '/' + '予実管理表.xlsx')
```

```python
#予実管理表の一番左の不要な列を削除する
#予実管理表の読み込み
workbook = openpyxl.load_workbook(export_file_path + '/予実管理表.xlsx')
#操作するシートを指定
worksheet = workbook.worksheets[0]
worksheet.delete_cols(1)
workbook.save(export_file_path + '/予実管理表_01.xlsx')
```
