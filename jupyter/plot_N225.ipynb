```python
from pandas_datareader import data
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
start = '2019-06-01'
end = '2020-06-01'

df = data.DataReader('^N225', 'yahoo', start, end)
```

```python
df.head(10)
```

```python
df.index
```

```python
date = df.index
#調整後の終値(Adj Close)をprice変数で定義
price = df['Adj Close']
```

```python
#グラフで可視化する（X軸：date、Y軸：price)
plt.plot(date, price)
```

```python
#グラフを大きくする
plt.figure(figsize=(30,10))
plt.plot(date, price)
```

```python
#線グラフに名前を付ける(plt.figureがないとグラフは小さくなる)
plt.figure(figsize=(30,10))
plt.plot(date, price,label='Nikkei225')
#ラベルを表示させる(legend)
plt.legend()
```

```python
plt.figure(figsize=(30,10))
plt.plot(date, price,label='Nikkei225')
#グラフにタイトルを付ける
plt.title('N225', color='blue', backgroundcolor='white', size=40, loc='center')
plt.legend()
```

```python
plt.figure(figsize=(30,10))
plt.plot(date, price,label='Nikkei225')
plt.title('N225', color='blue', backgroundcolor='white', size=40, loc='center')
#X軸、Y軸に名前を付ける
plt.xlabel('date', color='black', size=30)
plt.ylabel('price', color='black', size=30)
plt.legend()
```

```python
#移動平均の変数定義
#from curses import window


span01 = 5
span02 = 25
span03 = 50

#移動平均を算出するカラムを追加
df['sma01'] = price.rolling(window=span01).mean()
df['sma02'] = price.rolling(window=span02).mean()
df['sma03'] = price.rolling(window=span03).mean()
```

```python
#省略なしでheadを表示させる
pd.set_option('display.max_rows', None)
df.head(100)
```

```python
plt.figure(figsize=(30,10))
plt.plot(date, price,label='Nikkei225', color='#99b898')

plt.plot(date,df['sma01'], label='sma01', color='#e84a5f')
plt.plot(date,df['sma02'], label='sma02', color='#ff847c')
plt.plot(date,df['sma03'], label='sma03', color='#feceab')

plt.title('N225', color='blue', backgroundcolor='white', size=40, loc='center')
plt.xlabel('date', color='black', size=30)
plt.ylabel('price', color='black', size=30)
plt.legend()
```

```python
#東芝

start = '2019-06-01'
end = '2020-05-01'
company_code = '6502.JP'
```

```python
df = data.DataReader(company_code, 'stooq')
df = df[(df.index>= start) & (df.index<= end)]
date = df.index
price = df['Close']

span01 = 5
span02 = 25
span03 = 50

#移動平均を算出するカラムを追加
df['sma01'] = price.rolling(window=span01).mean()
df['sma02'] = price.rolling(window=span02).mean()
df['sma03'] = price.rolling(window=span03).mean()

plt.figure(figsize=(30, 15))
plt.subplot(2,1,1,)

plt.plot(date, price, label='Close', color='#99b898')
plt.plot(date, df['sma01'], label='sma01', color='#e84a5f')
plt.plot(date, df['sma02'], label='sma02', color='#ff847c')
plt.plot(date, df['sma03'], label='sma03', color='#feceab')
plt.legend()

plt.subplot(2,1,2)
plt.bar(date, df['Volume'], label='Volume', color='grey')
plt.legend()
```

```python
def company_stock(start, end, company_code):
  df = data.DataReader(company_code, 'stooq')
  df = df[(df.index>= start) & (df.index<= end)]
  
  date = df.index
  price = df['Close']

  span01 = 5
  span02 = 25
  span03 = 50

  #移動平均を算出するカラムを追加
  df['sma01'] = price.rolling(window=span01).mean()
  df['sma02'] = price.rolling(window=span02).mean()
  df['sma03'] = price.rolling(window=span03).mean()

  plt.figure(figsize=(20, 10))
  plt.subplot(2,1,1,)

  plt.plot(date, price, label='Close', color='#99b898')
  plt.plot(date, df['sma01'], label='sma01', color='#e84a5f')
  plt.plot(date, df['sma02'], label='sma02', color='#ff847c')
  plt.plot(date, df['sma03'], label='sma03', color='#feceab')
  plt.legend()

  plt.subplot(2,1,2)
  plt.bar(date, df['Volume'], label='Volume', color='grey')
  plt.legend()
```

```python
company_stock('2019-06-01', '2020-06-01', '6502.JP')
```

```python
company_stock('2017-01-01', '2020-06-01', '6502.JP')
```
